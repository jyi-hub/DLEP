```python
import pandas as pd
import numpy as np
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')

# 1.1 定义城市顺序（您提供的列表）

city_order = cities
# 注意：请确保这里放入完整的369个城市名称

# 1.2 建立城市名称到索引的映射
city_to_idx = {city: idx for idx, city in enumerate(city_order)}
n_cities = len(city_order)

# 1.3 加载迁徙规模数据（迁入和迁出规模指数）
print("正在加载迁徙规模数据...")
# 假设有两个文件：迁入规模-2023.xlsx 和 迁出规模-2023.xlsx
inflow_scale_df = pd.read_excel('迁入规模-2023.xlsx')
outflow_scale_df = pd.read_excel('迁出规模-2023.xlsx')

# 1.4 检查数据格式并创建规模矩阵
# 假设DataFrame的列是：城市代码, 城市, 20230101, 20230102, ...
date_columns = [col for col in inflow_scale_df.columns if col not in ['城市代码', '城市']]
date_range = [f"{date}" for date in date_columns]  # 例如 ['20230101', '20230102', ...]

# 创建规模矩阵：shape = (城市数, 天数)
inflow_scale_matrix = np.zeros((n_cities, len(date_columns)))
outflow_scale_matrix = np.zeros((n_cities, len(date_columns)))

# 填充规模矩阵
print("正在构建规模矩阵...")
for idx, city in enumerate(city_order):
    # 迁入规模
    inflow_row = inflow_scale_df[inflow_scale_df['城市'] == city]
    if not inflow_row.empty:
        inflow_scale_matrix[idx] = inflow_row[date_columns].values[0]
    
    # 迁出规模
    outflow_row = outflow_scale_df[outflow_scale_df['城市'] == city]
    if not outflow_row.empty:
        outflow_scale_matrix[idx] = outflow_row[date_columns].values[0]

print(n_cities, len(date_columns))# 查看城市数与日期数
print([c for c in city_order if c not in set(inflow_scale_df['城市'])])#检查未匹配城市
#检查矩阵形状与样例值： 
print(inflow_scale_matrix.shape, inflow_scale_matrix[:3,:5])
```

```python
import os

# 2.1 初始化MII和MOI矩阵
MII_matrix = np.zeros((n_cities, n_cities, len(date_columns)))
MOI_matrix = np.zeros((n_cities, n_cities, len(date_columns)))

print("正在计算MII和MOI矩阵...")

# 2.2 预读所有 OD 比例文件
inflow_dir = r"D:\GIT\DLEP\data\human_mobility_data\迁入"
outflow_dir = r"D:\GIT\DLEP\data\human_mobility_data\迁出"

inflow_ratio_dfs = {}
outflow_ratio_dfs = {}
missing_files = []

for city in city_order:
    p_in = os.path.join(inflow_dir, f"{city}-迁入数据.csv")
    p_out = os.path.join(outflow_dir, f"{city}-迁出数据.csv")
    if os.path.exists(p_in):
        inflow_ratio_dfs[city] = pd.read_csv(p_in, encoding='gbk')
    else:
        inflow_ratio_dfs[city] = None
        missing_files.append(p_in)
    if os.path.exists(p_out):
        outflow_ratio_dfs[city] = pd.read_csv(p_out, encoding='gbk')
    else:
        outflow_ratio_dfs[city] = None
        missing_files.append(p_out)

print(f"缺失 OD 文件数: {len(missing_files)}")

# 2.3 计算 MII 和 MOI
print("正在计算MII和MOI矩阵...")
for i, source_city in enumerate(tqdm(city_order, desc="处理源城市")):
    for j, target_city in enumerate(city_order):
        # ===== MII(i,j) = (j视角下从i的迁入比例) × ISI(j) × 1000 =====
        # 从 target_city(j) 的迁入文件中找 source_city(i)
        if inflow_ratio_dfs[target_city] is not None:
            row_mii = inflow_ratio_dfs[target_city][inflow_ratio_dfs[target_city]['城市'] == source_city]
            ipr_i_from_j_view = row_mii[date_columns].fillna(0).values[0] if not row_mii.empty else np.zeros(len(date_columns))
        else:
            ipr_i_from_j_view = np.zeros(len(date_columns))
        
        # ISI(j) = j 的迁入规模
        isi_j = inflow_scale_matrix[j, :]
        
        # MII(i,j) = IPR × ISI(j) × 1000
        MII_matrix[i, j, :] = ipr_i_from_j_view * isi_j * 1000
        
        # ===== MOI(i,j) = (j视角下到i的迁出比例) × OSI(j) × 1000 =====
        # 从 target_city(j) 的迁出文件中找 source_city(i)
        if outflow_ratio_dfs[target_city] is not None:
            row_moi = outflow_ratio_dfs[target_city][outflow_ratio_dfs[target_city]['城市'] == source_city]
            opr_to_i_from_j_view = row_moi[date_columns].fillna(0).values[0] if not row_moi.empty else np.zeros(len(date_columns))
        else:
            opr_to_i_from_j_view = np.zeros(len(date_columns))
        
        # OSI(j) = j 的迁出规模
        osi_j = outflow_scale_matrix[j, :]
        
        # MOI(i,j) = OPR × OSI(j) × 1000
        MOI_matrix[i, j, :] = opr_to_i_from_j_view * osi_j * 1000

print("MII和MOI矩阵计算完成！")

```

```python
# 3.1 计算每日的flow矩阵：flow_ij(t) = (MII_ij(t) + MOI_ij(t)) / 2
flow_daily = (MII_matrix + MOI_matrix) / 2
# 截取前305天的数据（假设数据有305天）
flow_daily = flow_daily[:,:,0:306]
# 3.2 计算时间平均值：对时间维度取平均
print("正在计算时间平均flow矩阵...")
flow_matrix = np.mean(flow_daily, axis=2)

# 3.3 将flow矩阵转换为DataFrame，按指定城市顺序
flow_df = pd.DataFrame(flow_matrix, index=city_order, columns=city_order)

print("Flow矩阵形状:", flow_df.shape)
print("Flow矩阵前5行前5列:")
print(flow_df.iloc[:5, :5])
```

```python
# 4.1 对flow矩阵进行Min-Max标准化
def min_max_normalize(matrix):
    """对矩阵进行Min-Max标准化到[0, 1]区间"""
    min_val = np.min(matrix)
    max_val = np.max(matrix)
    
    if max_val == min_val:  # 防止除以0
        return np.zeros_like(matrix)
    
    normalized = (matrix - min_val) / (max_val - min_val)
    return normalized

# 4.2 应用标准化
flow_matrix_normalized = min_max_normalize(flow_matrix)
flow_normalized_df = pd.DataFrame(flow_matrix_normalized, 
                                   index=city_order, 
                                   columns=city_order)

print("标准化后Flow矩阵统计:")
print(f"最小值: {np.min(flow_matrix_normalized):.6f}")
print(f"最大值: {np.max(flow_matrix_normalized):.6f}")
print(f"平均值: {np.mean(flow_matrix_normalized):.6f}")
print(flow_matrix_normalized)
```
